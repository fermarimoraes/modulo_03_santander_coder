```python
import pandas as pd
import unidecode
import numpy as np
```

```python
def separar_endereco(endereco):
    endereco = endereco.split(' ')
    numero = endereco.pop()
    endereco = ' '.join(endereco)
    return endereco + '|' + numero
```

```python
def tipo_escola(escola):
    tipo = ''
    if 'EM' in escola or 'E.M' in escola or 'ESCOLA MUNICIPAL' in escola:
        tipo = 'EM'
    elif 'CENTRO INTEGRADO' in escola or 'CIEP' in escola:
        tipo = 'CIEP'
    else:
        tipo = 'COLEGIO'
    return escola + '|' + tipo
```

```python
def truncate(numero):
    numero = float(numero.replace(',','.'))
    numero = numero * 100000
    numero = np.trunc(numero)
    numero = numero / 100000
    return numero
```

```python
df = pd.read_csv("../datasets/escolas.csv")
df = df.rename(columns={'Escolas_Postos':'escolas_postos', 'BAIRRO':'bairro', 'ENDEREÇO ':'endereço'})
df['escolas_postos'] = df['escolas_postos'].apply(unidecode.unidecode).apply(lambda str:str.upper())  
df['bairro'] = df['bairro'].apply(unidecode.unidecode).apply(lambda str:str.upper()) 
df['endereço'] = df['endereço'].apply(unidecode.unidecode).apply(lambda str:str.upper().replace('R.','RUA').replace('AV.', 'AVENIDA').replace('ESTR.', 'ESTRADA').replace('PCA.', 'PRACA')) 
df['lat'] = df['lat'].apply(truncate) 
df['lon'] = df['lon'].apply(truncate) 
df['id'] = df['id'].apply(lambda id: str(id).rjust(3,'0'))
df[['endereço','numero']] = df['endereço'].apply(separar_endereco).str.split('|', expand= True)
df[['escolas_postos','tipo_da_escola']] = df['escolas_postos'].apply(tipo_escola).str.split('|', expand= True)

df
```

```python
df.to_csv('../datasets/escolas_transformado.csv')
```
